```python
import random as rd 
import pandas as pd 
```

```python
genData = pd.read_excel("preferences.xlsx", header= None)
genData = genData.iloc[:4]
genDataBack = genData
```

```python
n = 200

genData = genDataBack.copy()

genData.loc[1][0] = n

rd.seed(None)

billDevVals = [i for i in range(0, 6) ] + ([0, 1, 2] * 2)

weatherDevVals = [i for i in range(-5, 6) ] + ([-1, 0, 1] * 2)

numRows = 0

for i in range(n):
     billDev = rd.choice(billDevVals)
     weatherDev = rd.choice(weatherDevVals)


     initialValues = [rd.randint(40, 240), rd.randint(0, 14), rd.randint(0, 14),
          rd.randint(0, 14), rd.randint(0, 14), rd.randint(0, 35), rd.randint(0, 14),
          billDev, weatherDev]

     tolerance = []

     while billDev > 2 or abs(weatherDev) > 1:
          if billDev > 2:
               billDev = rd.choice(billDevVals)

          if abs(weatherDev) > 1:
               weatherDev = rd.choice(weatherDevVals)

          lst1 = [None] * len(initialValues)
          lst1[-2] = billDev
          lst1[-1] = weatherDev
          tolerance.append(lst1)

     # genData.append(initialValues)
     genData.loc[len(genData)] = initialValues

     for row in tolerance:
          genData.loc[len(genData)] = row
     
     numRows += len(tolerance) + 1

genData.loc[0][1] = "Number of rows"
genData.loc[1][1] = numRows

# 240	0	14	14	14	30	7	2	-2
# print(genData)
genData
```

```python
pd.DataFrame.to_excel(genData, "preferences.xlsx", header = False, index = False)
```
